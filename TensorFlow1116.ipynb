```python
import tensorflow as tf
import numpy as np

#https://morvanzhou.github.io/tutorials/machine-learning/tensorflow/3-1-add-layer/
#定義一個添加層的函數(node)
#定义添加神经层的函数def add_layer(),
#它有四个参数：输入值、输入的空間、输出的空間和激励函数，
#我们设定默认的激励函数是None

def add_layer(inputs, input_tensors,
             output_tensors, activation_function = None):
    #隨機變數用常態分配，用input,output值建一個矩陣
    W = tf.Variable( tf.random_normal([input_tensors,output_tensors]))
    
    b = tf.Variable( tf.random_normal([1, output_tensors]) +0.1)
    #未激活的值
    formula = tf.matmul(inputs ,W) + b
    
    if activation_function is None:
        #不跑激活函數
        outputs = formula
    else:
        #如果有給激活涵數，就跑定義好的激活函數
        outputs = activation_function(formula)
    return outputs
    
#準備資料
x_data = np.random.rand(100)
#重新定義維度，把數列排程 nX1的矩陣
x_data = x_data.reshape(len(x_data),1)
y_data = x_data *0.1 + 0.3

#建立feeds
#tf.placeholder()表示暫存區(定義進來資料的型態,shape)
#shape=[None,1] None表示輸入有多少都可以，也是 nX1矩陣
x_feeds = tf.placeholder(tf.float32, shape = [None,1])
y_feeds = tf.placeholder(tf.float32, shape = [None,1])

#加一個隱藏層
hidden_layer = add_layer(x_feeds, input_tensors = 1,
                        output_tensors = 10, activation_function = None)
#加一個輸出層
output_layer = add_layer(hidden_layer, input_tensors = 10,
                        output_tensors = 1, activation_function = None)

#定義loss與要使用的Optimizer
loss = tf.reduce_mean(tf.square( y_feeds - output_layer))
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.01)
train = optimizer.minimize(loss)

#初始化Graph並開始運算
init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)

#x_feeds以x_data輸入; y_feeds以y_data輸入
for step in range(200):
    sess.run(train, feed_dict = {x_feeds: x_data,
                                y_feeds: y_data})
    if step % 20 == 0:
        print(sess.run(loss, feed_dict = {x_feeds: x_data,
                                          y_feeds: y_data
        }))
#loss 的數值都在降低，表示類似模型不斷在逼近真實模型


sess.close()
```

```python
import tensorflow as tf
import numpy as np

#https://morvanzhou.github.io/tutorials/machine-learning/tensorflow/3-1-add-layer/
#定義一個添加層的函數(node)
#定义添加神经层的函数def add_layer(),
#它有四个参数：输入值、输入的空間、输出的空間和激励函数，
#我们设定默认的激励函数是None

def add_layer(inputs, input_tensors,
             output_tensors, activation_function = None):
        
    with tf.name_scope('Layer'):
        with tf.name_scope('Weights'):
            W = tf.Variable(tf.random_normal([input_tensors, output_tensors]))
        
        with tf.name_scope('Biases'):
            b = tf.Variable(tf.zeros([1, output_tensors]))
        
        with tf.name_scope('Formula'):
            formula = tf.add(tf.matmul(inputs, W), b)
        
        if activation_function is None:
            outputs = formula
        else:
            outputs = activation_function(formula)
        return outputs
    
#準備資料
x_data = np.random.rand(100)
#重新定義維度，把數列排程 nX1的矩陣
x_data = x_data.reshape(len(x_data),1)
y_data = x_data *0.1 + 0.3

#建立feeds
#tf.placeholder()表示暫存區(定義進來資料的型態,shape)
#shape=[None,1] None表示輸入有多少都可以，也是 nX1矩陣
x_feeds = tf.placeholder(tf.float32, shape = [None,1])
y_feeds = tf.placeholder(tf.float32, shape = [None,1])

#加一個隱藏層
hidden_layer = add_layer(x_feeds, input_tensors = 1,
                        output_tensors = 10, activation_function = None)
#加一個輸出層
output_layer = add_layer(hidden_layer, input_tensors = 10,
                        output_tensors = 1, activation_function = None)

#定義loss與要使用的Optimizer
loss = tf.reduce_mean(tf.square( y_feeds - output_layer))
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.01)
train = optimizer.minimize(loss)

#初始化Graph並開始運算
init = tf.global_variables_initializer()
sess = tf.Session()
#sess.run(init)


#將視覺化輸出
writer = tf.summary.FileWriter("TensorBoard/",
                              graph = sess.graph)

#開始運算
sess.run(init)


#x_feeds以x_data輸入; y_feeds以y_data輸入
for step in range(200):
    sess.run(train, feed_dict = {x_feeds: x_data,
                                y_feeds: y_data})
    
    
#    if step % 20 == 0:
#        print(sess.run(loss, feed_dict = {x_feeds: x_data,
#                                          y_feeds: y_data
#        }))


#loss 的數值都在降低，表示類似模型不斷在逼近真實模型




sess.close()
```
